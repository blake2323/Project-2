```python
beer_businesses = []

for business in data['businesses']:
    name = business['name']
    coordinates = business['coordinates']
    rating = business['rating']
    review_count = business['review_count']
    beer_businesses.append({"Name": name, "Coordinates": coordinates, "Rating": rating, 
                            "Review_Count": review_count})
```

```python
print(beer_businesses)

    
```

```python
beer_businesses_df = pd.DataFrame(beer_businesses)
beer_businesses_df
```

```python
coords = beer_businesses_df['Coordinates']
coords
```

```python
lats = []
longs = []

for coord in coords:
    
    lat = coord['latitude']
    long = coord['longitude']
    lats.append(lat)
    longs.append(long)

print(lats)
```

```python
beer_businesses_df['lat'] = lats
beer_businesses_df['long'] = longs
```
